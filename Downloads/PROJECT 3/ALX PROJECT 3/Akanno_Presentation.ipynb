### Investigation Overview

> The present study on the exploration of prosper loan dataset aims to identify the various features that affect loan status.The data was initially assessed to understand and gain more insight on the dataset and was afterwards cleaned to produce a quality data. Importantly, some columns that are not relevant were removed and datatype was changed to a proper datatype.
Furthermore, the data was visualised using univariate, bivariate, and multivariate appraoch which provided more insights from the dataset. 
> Based on the findings, the number of completed loan repayment is higher than the non-remittance loan repayment. The employed had the highest completed loan status and surprisingly, the highest number of non-remittance loan status. While some positive correlations were observed between features, some features were negatively correlated. In addition , there is no significant relationship between RV and green loans with either the completed and non-remittance loan status. Overall, borrowers with higher BorrowerRates are more likely to default.

### Status of Loan Repayment
> The plot shows the freuency of completed and non-remittance loan in the Prosper Loan Dataset.

>The plot indicates that the number of completed loan is more than the non-remittance, whcih is approximately 20000 and 5000 respectively.

### Distribution of Loan Status by Listing Category

> The plot shows the relationship between the loan status and the listing category.

> Debt consolidation has the highest completed and non-remittance loan repayment. This shows a very high correlation between debt consolidation and completed loan as well as non-remittance loan status. Meanwhile RV and Green Loans have no correlation with either completed loan or non-remittance loan status.

### Loan Status by Income Range and Loan Amount

> This plot shows the relationship between loan status, income range and loan amount

> It can be indicated that income range of 100,000 + dollars has the highest completed and non-remittance loan status as well as the loan original amount. It can also be observed that income range between 1-24,999 dollars has the lowest loan status and loan original amount
