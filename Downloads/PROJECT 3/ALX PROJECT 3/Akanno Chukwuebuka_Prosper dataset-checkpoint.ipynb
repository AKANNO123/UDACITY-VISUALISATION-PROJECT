### Data Cleaning

>The following quality and tidiness issues were discovered in the analysis

1. The loanStatus should be either Completed and non remittance as status
2. The Loan Origination format should be datetime datatype
3. The information in the listing category column should be decoded to have the same information in the data dictionary which is easier to capture and understand.
4. Renaming of some columns such as Listing Category (Numeric) and Prosper Rating(Alpha).
5. Some of the columns in the dataset have missing values

Define: standardize LoanStatus into completed and non-remittance varibales

Code:

```python
def standardize_loan_status2(row):
    if row['LoanStatus'] not in ['Current', 'Completed', 'FinalRepaymentInProgress']:
        return 'non_remittance'
    else:
            return row['LoanStatus']
```

```python
df_copy = df.copy()
df_copy['LoanStatus'] = df_copy.apply(standardize_loan_status2, axis=1)
```

```python
df_copy['LoanStatus'].value_counts()
```

```python
df_copy = df_copy[df_copy.LoanStatus != 'Current']
```

# Univariate Exploration

1. How many completed and non completed loan were recorded?

```python
def makeplot(col, start = None, stop = None):
    if start or stop != None:
    
        # Return the Series having unique values
        x = df_clean_copy[col].unique()[start:stop]

        # Return the Series having frequency count of each unique value
        y = df_clean_copy[col].value_counts()[start:stop]

    else:
        x = df_clean_copy[col].unique()
        y = df_clean_copy[col].value_counts()
        
    plt.bar(x,y)

    # Labeling the axes
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=90)
    plt.title(col + ' and loan Repayment')

    # Dsiplay the plot
    return plt.show()
```

```python
makeplot('LoanStatus')
```

> The plot indicates that the number of completed loan is more than the non-remittance, whcih is approximately 20000 and 5000 respectively.

```python
#IncomeRange and ProsperRating as ordered categorical variable
ordinal_var_dict = {'IncomeRange': ['$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+'],
                    'ProsperRating': ['A', 'AA', 'B', 'C', 'D', 'E', 'HR']}
                                   
for var in ordinal_var_dict:
    ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                                categories = ordinal_var_dict[var])
    df_clean_copy[var] = df_clean_copy[var].astype(ordered_var)
```
